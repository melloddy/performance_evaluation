```python
import pandas as pd
```

```python
# Load Excel file (the files need to be downloaded next to this script)
file = 'all_compute_plans_melloddy.csv'
df = pd.read_csv(file, delimiter=',', low_memory=False)
df = df.dropna(subset=['name'])
```

```python
def filter_df(df, phase=[], dataset=[], model_type=[], partner=[]):
    """
    Filters the dataframe of compute plans, based on the the phase, dataset, model_type and partner
    AND filtering is performed between parameters
    OR filtering is performed within parameters
    
    Args
        df (pandas.DataFrame): data to filter
        phase, dataset, model_type, partner (lists): values to filter on
    
    Returns
        df_filtered (pandas.DataFrame): filtered data
    """
    
    df_filtered = df.copy()
    
    # words to keep in the column "name"
    filters_name = [phase, dataset, model_type]
    for words in filters_name: 
        df_filtered = df_filtered[df_filtered.name.str.contains('|'.join(words))]
        
    # words to keep in the column "worker"
    filters_worker = [partner]
    for words in filters_worker: 
        df_filtered = df_filtered[df_filtered.worker.str.contains('|'.join(words))]
    
    return df_filtered
```

```python
phase = ['PH1'] # 'PH1', 'PH2', 'PH3'
dataset = ['classification'] # 'classification', 'regression', 'clsaux', 'hybrid'
model_type = ['MP'] # 'SP', 'MP'
partner = ['pharma1'] # 'pharma1', 'pharma2', 'pharma3', 'pharma4','pharma5', 'pharma6', 'pharma7', 'pharma8', 'pharma9', 'pharma10'

df_filtered = filter_df(df, phase, dataset, model_type, partner)
df_filtered.head()
```
